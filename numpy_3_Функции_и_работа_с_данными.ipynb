###Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

````python
ratings = pd.read_csv('ratings.csv')
ratings
````

````python
def films_class(param):
    """классифицирует фильмы из материалов занятия по следующим правилам:
        оценка 2 и меньше - низкий рейтинг
        оценка 4 и меньше - средний рейтинг
        оценка 4.5 и 5 - высокий рейтинг"""
    if param <= 2:
      return 'низкий рейтинг'
    elif param <= 4:
      return 'средний рейтинг'
    else:
      return 'высокий рейтинг'
````

````python
?films_class
````

````python
ratings['class'] = ratings['rating'].apply(films_class)
ratings.query('rating == 3').head()  #проверка)
````

````python
# ещё одна проверка))
high_class= ratings[ratings['rating'].isin([4, 4.5, 5])]
high_class
````

###Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл **movies.csv** и **ratings.csv** из базы https://grouplens.org/datasets/movielens/). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.


При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

  * для каждой строки пройдите по всем годам списка years

  * если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска

  * если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга


PS. все файлы находятся в архиве «дополнительные файлы к лекции и дз»

````python
import numpy as np
````

````python
years = np.arange(1950, 2011)
````

````python
years
````

````python
movies = pd.read_csv('movies.csv')
movies
````

````python
#проверяю на всякий случай тип года, чтоб знать в каком типе возвращать 1900
for year in years:
  print(type(year))
  break
````

````python
def production_year(film):
  #  *для каждой строки пройдите по всем годам списка years
  for year in years:
    #  *если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    if str(year) in film:
      return year
  
  # *если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
  return 1900
````

````python
movies['year'] = movies['title'].apply(production_year)
movies
````

````python
all_info = movies.set_index('movieId').join(ratings.set_index('movieId'))	
all_info
````

````python
show_rat = all_info.groupby('year').mean('rating').sort_values(by="rating", ascending=False)
show_rat
````
