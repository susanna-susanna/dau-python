###Задание 1
Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

*   если источник traffic_source равен yandex или google, то в source_type ставится organic
*   для источников paid и email из России - ставим ad
*   для источников paid и email не из России - ставим other
*   все остальные варианты берем из traffic_source без изменений


````python
import pandas as pd
````

````python
log = pd.read_csv('visit_log.csv', sep=';')
log
````

````python
def complex_filter(row):
    """Фильтрует строки по следующим правилам:
        если источник traffic_source равен yandex или google, то в source_type ставится organic
        для источников paid и email из России - ставим ad
        для источников paid и email не из России - ставим other
        все остальные варианты берем из traffic_source без изменений
        """
    traf_source = row['traffic_source']
    region = row['region']
    
    if traf_source == 'yandex' or traf_source == 'google':
        return 'organic'
    elif (region == 'Russia') and (traf_source == 'paid' or traf_source == 'email'):
        return 'ad'
    elif (region != 'Russia') and (traf_source == 'paid' or traf_source == 'email'):
        return 'other'
        
    return traf_source
````

В первом задании можно также использовать вариант решения через метод loc (обычно работает быстрее). Например, для первого правила:
```
log.loc[log.traffic_source.isin(['yandex', 'google']), 'traffic_type'] = 'organic'
```

````python
log['source_type'] = log.apply(complex_filter, axis=1)
log
````


###Задание 3
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.


````python
ratings = pd.read_csv('ratings.csv')
ratings
````

````python
# группируем по юзерам
# считаем их активности и даты первой и последней активностей
group_user = ratings.groupby('userId').agg({'userId':'count','timestamp':['min','max']})
# group_user = ratings.groupby('userId').agg({'movieId':'count','timestamp':['min','max']}) #т.к нам нужно посчитать кол-во строк, нет разницы, по какому параметру вести счёт count
group_user
````

````python
# переименую столбец для удобства понимания)))
group_user.rename({'userId': 'activity', 'count': 'level'}, axis=1, inplace=True)
group_user
````

````python
# отметка для себя: если не правильно прописана вложенность -- будет ошибки! 
# Поэтому внимательно. Оставлю, чтоб всегда было перед глазами
active_user = group_user.loc[group_user.activity > 100]
active_user
````

````python
# отметка для себя: если не правильно прописана вложенность -- будет ошибки!
# Поэтому внимательно. Оставлю, чтоб всегда было перед глазами
active_user = group_user.loc[group_user.level > 100]
active_user
````

````python
#нужно указать правильно вложенность при названии столбца
active_user = group_user.loc[group_user.activity.level > 100]
active_user
````

````python
# теперь ищем LTV life time value
active_user['LTV'] = active_user.timestamp.max - active_user.timestamp.min
````

````python
# пробую такой записью найти разницу между max и min
active_user['LTV'] = active_user.timestamp['max'] - active_user.timestamp['min']
````

````python
# среднее LTV
active_user['LTV'].mean()
````

### К домашнему заданию, задача 4
Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

````python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
````

````python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
````

````python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
````

````python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
````

````python
rzd_auto = rzd.merge(auto, on='client_id', how='outer')
rzd_auto
````

````python
rzd_auto_air = rzd_auto.merge(air, on='client_id', how='outer')
rzd_auto_air
````
